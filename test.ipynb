```julia
using Plots
```

```julia
# Define the function
function f(x)
    return x^2 + 2x + 1
end
# Generate x values
x = -10:0.1:10
# Compute y values
y = f.(x)

; # Suppresses the output
```

```julia
# Create the plot
scatter(x, y, label="f(x) = x^2 + 2x + 1", title="Plot of f(x)", xlabel="x", ylabel="f(x)", legend=:topright)
raw"x-axis"
ylabel!("y-axis")
title!("Plot of f(x)")
```

```julia
# random walk 
using Random
function random_walk(n)
    steps = rand([-1, 1], n)  # Generate n random steps of -1 or 1
    return cumsum(steps)       # Cumulative sum to get the position
end
# Generate random walk data
n = 100000
walk = random_walk(n)
# Create the random walk plot
plot(walk, label="Random Walk", title="Random Walk Simulation", xlabel="Step", ylabel="Position", legend=:topright)
# Add grid lines
# grid!(true)
# Add a title
title!("Random Walk Simulation")
```
